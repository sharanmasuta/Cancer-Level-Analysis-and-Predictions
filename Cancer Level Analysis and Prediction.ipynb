# Pie graph

```python
def gender():
    
    
    g=df['Gender'].value_counts()
    gender=[g[1],g[2]]
    plt.figure(figsize=(8,6))
    mycolors = ['b' ,"hotpink", "#4CAF50"]
    label=("Male","Female")
    textprops = {"fontsize":13 , 'fontweight':'bold'}
    wedgeprops = {"linewidth":2 , "width":1,"edgecolor":"k" }
    
    plt.pie(gender,labels=label,colors=mycolors,autopct="% .3f %%",explode=(.03,.04),shadow=True,radius=1,startangle=250,textprops = textprops,wedgeprops=wedgeprops)#frame=True)
    plt.title("Male vs Female",fontweight ='bold', fontsize = 20)
    plt.legend(title="Gender")
    plt.axis("equal")
    
gender()
```

```python
data_pie = df['Level'].value_counts().rename_axis('Level').reset_index(name='patients_count')
```

```python
def level_cancer():
    plt.figure(figsize=(10,6))
    textprops = {"fontsize":15 , 'fontweight':'bold'}
    wedgeprops = {"linewidth":2 , "width":1,"edgecolor":"blue" }
    mycolors = ["red", "orange" ,"green"]
    legends=['High','Medium','Low']
    
    plt.pie(data_pie.patients_count , labels=legends,colors=mycolors,autopct = '%.1F%%',explode=(.01,.03,.06),shadow=True,radius=1,startangle=250,textprops = textprops,wedgeprops=wedgeprops)
    plt.title('Levels of the patients suffering from cancer',fontweight ='bold', fontsize = 20)
    plt.legend(title='Level')
    
level_cancer()   
    
  
```

# Machine learning

```python
# import algorithms
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn import model_selection
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
X=df[['Age','Gender','Air Pollution','Alcohol use','chronic Lung Disease','Smoking','Chest Pain', 'Coughing of Blood']]
Y=df['Level']

P=[[34,0,4,5,6,3,4,5]]

clf = DecisionTreeClassifier()
clf = clf.fit(X,Y)

print ("\n1) Using Decision Tree Prediction is :" + str(clf.predict(P)))

clf1=KNeighborsClassifier()
clf1=clf1.fit(X,Y)
print("2) using Kneighbors Classifier is :" +str(clf1.predict(P)))

clf2=RandomForestClassifier()
clf2=clf2.fit(X,Y)
print("3) using RandomForestClassifier is :" +str(clf2.predict(P)))

clf3=LogisticRegression()
clf3=clf3.fit(X,Y)
print("4) using Logistic Regression is :" +str(clf3.predict(P)))

```

```python
# for machine learning convert string to int value
def convert_to_int(val):
    val=val.replace("P","")
    return int(val)
```

```python
df['Patient Id'] = df['Patient Id'].apply(convert_to_int)
df.head()
```

# split out validation dataset

```python
cancerarray = df.values
X=cancerarray[:,:-1]
Y=cancerarray[:,-1]
Xtrain,Xvalidation,Ytrain,Yvalidation=model_selection.train_test_split(X,Y,test_size=0.20, random_state=9)
type(Xtrain)
model=KNeighborsClassifier()
model=model.fit(Xtrain,Ytrain)
predict=model.predict(Xvalidation)
print('Predictions : ',predict)
```

# To check the accuracy

```python
model=[]
model.append(DecisionTreeClassifier())
model.append(RandomForestClassifier())
model.append(KNeighborsClassifier())
model.append(svm.SVC(kernel='rbf'))
model.append(svm.SVC(kernel='linear'))
#model.append(LogisticRegression())

print("models = : " , model)

for i in model:
    Model=i
    Model=Model.fit(Xtrain,Ytrain)
    Prediction= Model.predict(Xvalidation)
    print('\n Accuracy by using', i ,'is : ',accuracy_score(Prediction,Yvalidation))
    print('Confusion Matrix = \n\n '  , confusion_matrix(Yvalidation, Prediction)) 
    print("Classification Report :\n",classification_report(Yvalidation, Prediction))    
```

# Kfold and cross val score

```python
for i in model:
    
    scoring = 'accuracy'
    kfold = model_selection.KFold(n_splits=10)
    cross_val = model_selection.cross_val_score(i, Xtrain, Ytrain, cv=kfold, scoring=scoring).mean()
    print(i,cross_val)   
```
